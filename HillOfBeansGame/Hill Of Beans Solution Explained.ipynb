## The Game
Consider a game in which two players compete over a hill of beans.  The players take turns removing a number of beans from the hill.  On any one turn, a player must take at least one, and no more than three beans.  When a player removes a bean from the hill, he or she places it in the bean receptacle shared by both players.  The player who removes the last bean or beans gets to keep all of the beans.

## Background
Reading the above premis, we can get a pretty thorough list of the rules of the game:


*   There are two players
*   Turns are alternating
*   Players must take at least one bean but not more than 3 or the number remaining, whichever is lower
*   The player that takes the last bean wins the game

Some things to remember while we're working:

*   We can start with either a set number of beans or specify at runtime
*   We don't necessarily know who is playing but we could figure it out at a later time

## The Setup
We're going to start with what we know we need before we actually play the game:

*   How many beans there are to start out with
*   The minimum number that the player should take
*   The maximum number that the player shoudl take



```python
totalBeansLeft = 15
maxBeansToTake = 3
minBeansToTake = 1
numberOfTurns = 0
```

## Defining the Turn and Coding the Game
We already know how a single turn goes:


1.   Player looks at how many beans there are
2.   Player decides to take "X" number of beans from the pile, making sure they're taking the correct number
3.   The turn passes to the next player

We can encompass all of these steps into a single loop that will iterate until there are no more beans left to take.



```python
while totalBeansLeft > 0:
  print("There are %d beans left..." %totalBeansLeft)
  
  numToTake = int(input("How many do you want to take? "))
  
  if(numToTake > totalBeansLeft) or (numToTake > maxBeansToTake):
    print("You can't take that many.... try again...")
  elif numToTake < minBeansToTake:
    print("You have to take at least %d beans..." % minBeansToTake)
  else:
    totalBeansLeft -= numToTake
    numberOfTurns += 1
    
print("The game has been won after %d turns!" % numberOfTurns)
```

## Incorporating "Artificial Intelligence"
Artificial Intelligence, or AI, can be found in a lot of the things that we interact with today. Obvious exampls are playing computer games against computer-controlled players but there's also AI found in things like cars, trains, logistics companies, etc.

For this game, we can actually create a simple AI that can take turns based off of the algorithm that we provide it. This can range to just taking a random amount within the game bounds up to actually giving it a "winning" algorithm to make it harder to beat. This example will give a method definition for a smarter opponent.

```python
def otherSmartTurn():
  aiTake = totalBeansLeft%(maxBeansToTake + 1)
  
  if aiTake == 0:
    aiTake += 1
    
  print("Other player takes %d beans..." % aiTake)
  return aiTake
```

Now, if we rerun the piece of code in "The Setup" to reset the game, we can now code a single-player version of the game against our new AI.

```python
while totalBeansLeft > 0:
  print("There are %d beans left..." %totalBeansLeft)
  
  numToTake = int(input("How many do you want to take? "))
  
  if(numToTake > totalBeansLeft) or (numToTake > maxBeansToTake):
    print("You can't take that many.... try again...")
  elif numToTake < minBeansToTake:
    print("You have to take at least %d beans..." % minBeansToTake)
  else:
    totalBeansLeft -= numToTake
    numberOfTurns += 1
    
    if totalBeansLeft > 0:
      totalBeansLeft -= otherSmartTurn()
      numberOfTurns += 1
    
print("The game has been won after %d turns!" % numberOfTurns)
```
